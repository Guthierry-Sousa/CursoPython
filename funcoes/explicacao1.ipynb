2. Tipos de par√¢metros

```python
# Posicionais

def soma(a,b):
    return a+b

r = soma(2,5)

print(r)
```

```python
# Com valor padr√£o (default)

def norma_euclidiana(x, y, z = 0):
    return (x**2 + y**2 + z**2)**0.5

n1 = norma_euclidiana(3, 4)
print(n1)

n2 = norma_euclidiana(3, 4, 12)
print(n2)
```

```python
# Nomeados

def cadastrar(nome, sobrenome):
    print(f"Usu√°rio com dados - Nome: {nome.upper()} | Sobrenome: {sobrenome.upper()}, foi cadastrado!")

# Passando da forma posicional
cadastrar('guthierry', 'sousa')

# Passando da forma nomeado (clareza sem√¢ntica e evita erros)
cadastrar(sobrenome = 'sousa', nome = 'guthierry')
```
