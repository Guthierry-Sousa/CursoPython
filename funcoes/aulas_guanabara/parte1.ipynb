```python
# Exercício 1: Faça um programa que tenha uma função chamada que receba as dimensões de um terreno retangular (largura e comprimento) e mostre a área do terreno.

def calcular_area(largura, comprimento):

    area_terreno = largura * comprimento

    print(f"Lagura = {largura} | Comprimento = {comprimento}\n")
    print(f"Área = {area_terreno:.2f} m^2")

l = float(input("Informe a largura do terreno(m): "))
c = float(input("Informe o comprimento do terreno(m): "))

calcular_area(l,c)
```

```python
# Exercício 2: Faça um programa que tenha uma função que receba um texto qualquer como parâmetro e moestre uma mensagem com tamanho adaptável.

def escrever_mensagem(mensagem):

    tam_msg = len(mensagem)

    print("-"*tam_msg)
    print(mensagem)
    print("-"*tam_msg)

escrever_mensagem("Amo a elis")
escrever_mensagem("Adoro estudar python e dados")
```

```python
# Exercício 3: Faça um programa que tenha uma função que receba três parâmetros: início, fim e passo e realize a contagem. Seu programa tem que realizar três contagens através da função criada.
import time


def realizar_contagem(inicio, fim, passo = 1):

    if passo == 0:
        
            passo = 1

    if inicio > fim and passo > 0:

        passo = -passo


    print(f"Iniciando contagem de {inicio} até {fim} com passo de {abs(passo)}.")

    for i in range(inicio, fim+passo, passo):
        time.sleep(0.5)
        print(i, end = " ")

    print("FIM da contagem.")


realizar_contagem(1,10,1)
realizar_contagem(10,2,2)

i = int(input("Informe o inicio da contagem: "))
f = int(input("Informe o fim da contagem: "))
p = int(input("Informe o passo da contagem: "))

realizar_contagem(i, f, p)

```

```python
# Exercício 4: Faça um programa que tenha uma função que receba vários parâmetros com valores inteiros. Seu programa tem que analisar todos os valores e dizer qual deles é o maior.

def maior_valor(*n):

    maior = max(n)

    n = list(map(str, n))

    print(f"Os valores informados foram: {" , ".join(n)} .")
    print(f"O maior valor passado foi {maior}")

maior_valor(0)
maior_valor(2,3,4,1,10)
```

```python
# Exercício 5: Faça um programa que tenha uma lista e duas funções. A primeira função vai sortear 5 números e vai colocó-los dentro da lista e a segunda função vai mostrar a soma entre todos os valores pares sorteados pela função anterior.
import random


def sortear_numeros():

    numeros_sorteados = [random.randint(1,1000) for i in range(5)]

    print(f"Os números sorteados foram: {numeros_sorteados}")

    return numeros_sorteados


def somar_numeros_pares(numeros):

    soma = 0

    numeros_pares = []

    for n in numeros:

        if n % 2 == 0:

            numeros_pares.append(n)

            soma += n

    print(f"Os números pares foram: {numeros_pares}\n")
    print(f"Somatório = {soma}")

    

lista_numeros = sortear_numeros()

somar_numeros_pares(lista_numeros)
```

```python
# Exercício de fixação - Calcule o fatorial de um número, retorne e exiba na tela.

def calcular_fatorial(n):

    fat = 1

    for i in range(1,n+1):

        fat *= i

    return fat

num = int(input("Informe um número: "))

fatorial = calcular_fatorial(num)

print(f"O fatorial de {num} é: {fatorial}!")

```

```python
# Exercício 6: Crie um programa que tenha uma função que vai receber como parâmetro o ano de nascimento de uma pessoa, retornando um valor literal indicando se uma pessoa tem voto NEGADO, OPCIONAL ou OBRIGATÓRIO nas eleições.

def verificar_voto(ano_nascimento):

    from datetime import date

    idade = date.today().year - ano_nascimento

    if 18 <= idade < 65:

        return f"Com {idade} anos: VOTO OBRIGATÓRIO"
    
    elif idade >= 16 or idade >= 65:

        return f"Com {idade} anos: VOTO OPCIONAL"
    
    else:

        return f"Com {idade} anos: VOTO NEGADO"
    
ano_nasci = int(input("Informe sua idade: "))

resultado = verificar_voto(ano_nasci)

print(resultado)

```

```python
# Exercício 7: Crie um programa que tenha uma função que receba dois parâmetros: o primeiro indica o número a calcular e o outro será um valor lógico(opcional) indicando se será mostrado ou não na tela o processo de cálculo do fatorial.

def calcular_fatorial(n, show = False):

    """
        -> Calcula o Fatorial de um número
        :parâmetro n: O número a ser calculado o fatorial
        :parâmetro show: (opcional) Mostrar ou não a conta realizada
        :return: O fatorial do valor n
    """

    fat = 1

    for i in range(n,0,-1):

        fat *= i

    if show == False:

        return fat
    
    else: 

        nums = map(str, range(n,0,-1))

        return f"{" x ".join(nums)} = {fat}"
    
print(calcular_fatorial(5))
print(calcular_fatorial(7,True))
help(calcular_fatorial)
```

```python
# Exercício 8: Faça um programa que tenha uma função, que receba dois parâmetros opcionais: o nome de um jogador e quantos gols ele marcou. O programa deverá ser capaz de mostrar a ficha do jogador, mesmo que algum dado não tenha sido informado corretamente.

def mostrar_ficha(nome='<desconhecido>', qtd_gols=0):
    print("-------------------------------------------")
    print(f"O jogador {nome} fez {qtd_gols} gol(s) no campeonato.")


name = input("Nome do jogador: ").strip()
gols = input("Número de gol(s): ").strip()

if name == "":
    name = '<desconhecido>'

if gols.isnumeric():
    gols = int(gols)
else:
    gols = 0

mostrar_ficha(nome=name, qtd_gols=gols)


```

```python
#Exercício 9: Crie um programa que tenha uma função que vai funcionar de forma semelhante à função input() do Python, só que fazendo a validação para aceitar apenas um valor numérico.

def ler_numero(txt):

    r = input(txt)

    while not (r.isnumeric()):

        r = input(txt)

    return r

n = ler_numero("Digite um número: ")

print(f"Você acabou de digitar o número: {n}")
```

```python
# Exercício 10: Faça um programa que tenha uma função que pode receber vários parâmetros e vai retornar um dicionário com as seguintes informações: Quantidade de notas, maior nota, menor nota, media da turma e a situação (opcional)

def calcular_notas(*n, situcao = False):

    """
        -> Calcula dados de uma turma a partir das notas
        :parâmetro n: Notas dos alunos
        :parâmetro situacao: (opcional) Mostrar ou não a situcao da turma
        :return: O dicionário com as informações calculadas
    """
    
    dados_notas = {}

    qtd_notas = len(n)
    dados_notas["Quantidade"] = qtd_notas

    max_value = max(n)
    dados_notas["Maior nota"] = max_value

    min_value = min(n)
    dados_notas["Menor nota"] = min_value

    media = sum(n)/len(n)
    dados_notas["Média"] = media

    if situcao:

        if media > 7:
            dados_notas["Situação"] = "BOA"
        elif media > 5:
            dados_notas["Situação"] = "RAZOÁVEL"
        else:
            dados_notas["Situação"] = "RUIM"

    return dados_notas

print(calcular_notas(10,6.5,10,4,7,8,9))
print(calcular_notas(10,6.5,10,4,7,8,9, situcao = True))
print(calcular_notas(10,6.5,10,0,6,6,6, situcao = True))


help(calcular_notas)


```

```python
# Exercício 11: Faça um mini-sistema que utiliza o Interactive Help. O usuário vai digitar o comando e o manual vai aparacer. Quando o usuário digitar a palavra 'FIM' o programa se encerrará.
import time

def py_help(funcao):
    help(funcao)

def exibir_menu():
    print("--------------- PyHelp -----------------")
    r = input("Informe o nome da função: ").strip()
    print("----------------------------------------")
    return r

while True:
    resultado = exibir_menu()
    
    if resultado.upper() == 'FIM':
        break

    py_help(resultado)
    time.sleep(0.5)

print("Programa finalizado. ATÉ LOGO.")
```
