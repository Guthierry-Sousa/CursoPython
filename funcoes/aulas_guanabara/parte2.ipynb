##### Docstrings

```python
import time
def realizar_contagem(inicio, fim, passo = 1):
    """
        ->Realiza a contagem e exibi na tela

        :parâmetro inicio: inicio da contagem
        :parâmetro fim: fim da contagem
        :parâmetro passo: passo da contagem (default = 1)

        :return: sem retorno

    """

    if passo == 0:
        
            passo = 1

    if inicio > fim and passo > 0:

        passo = -passo


    print(f"Iniciando contagem de {inicio} até {fim} com passo de {abs(passo)}.")

    for i in range(inicio, fim+passo, passo):
        time.sleep(0.5)
        print(i, end = " ")

    print("FIM da contagem.")

realizar_contagem(100,0,10)
help(realizar_contagem)
```
