```python
# Aplicação

def dividir_treino_teste(lista1, lista2, percentual_divisao):

    fator1 = int((1-(percentual_divisao/100)) * len(lista1))

    lista_teste1 = lista1[fator1:].copy()
    lista_treino1 = lista1[:fator1].copy()

    fator2 = int((1-(percentual_divisao/100)) * len(lista2))

    lista_teste2 = lista2[fator2:].copy()
    lista_treino2 = lista2[:fator2].copy()

    return lista_teste1, lista_teste2, lista_treino1, lista_treino2

precos_imoveis = [2.17,1.54,1.45,1.94,2.37,2.3,1.79,1.8,2.25,1.37]
tamanho_imoveis = [207, 148, 130, 203, 257, 228, 160, 194, 232, 147]

```

```python
precos_imoveis_teste, tamanho_imoveis_teste, precos_imoveis_treino, tamanho_imoveis_treino = dividir_treino_teste(precos_imoveis, tamanho_imoveis, 20)

print(f"\nPreços dos imóveis - Lista Treino: {precos_imoveis_treino} || Lista Teste: {precos_imoveis_teste}\n")

print(f"\nTamanho dos imóveis - Lista Treino: {tamanho_imoveis_treino} || Lista Teste: {tamanho_imoveis_teste}\n")
```
