```python
import math as mt
```

```python
def integrar_numericamente(f, a, b, n = 1000):

    delta_x = (b-a)/n
    soma = 0.0

    for i in range(n):

        x_i = a + i*delta_x
        x_next = a + (i+1)*delta_x

        soma += (f(x_i) + f(x_next))/2

    return soma * delta_x

def f(x):

    return mt.sin(x)

s = integrar_numericamente(f, 0, mt.pi)

print(f"Integral numérica de sin(x) em [0 , pi] = {s:.2f}")
        
```

```python
def preencher_matriz():

    n = int(input("Informe o número de linhas: "))
    m = int(input("Informe o número de colunas: "))

    matriz = []

    for i in range(n):

        linha = []

        for j in range(m):

            elemento = int(input(f"Informe o elemneto ({i}x{j}): "))

            linha.append(elemento)

        matriz.append(linha)

    return matriz

def preencher_vetor():

    n = int(input("Informe o tamanho do vetor: "))
    vetor = []

    for i in range(n):
        vetor.append(int(input(f"Informe o elemento {i}: ")))

    return vetor


def multiplicar_matriz_vetor(matriz, vetor):

    n = len(matriz)
    m = len(matriz[0])

    if len(vetor) != m:
        raise ValueError("Dimensões incompatíveis")

    resultado = []

    for i in range(n):
        soma = 0
        for j in range(m):
            soma += matriz[i][j] * vetor[j]
        resultado.append(soma)

    return resultado

mat = preencher_matriz()

vet = preencher_vetor()

r = multiplicar_matriz_vetor(mat, vet)

print(r)

```

```python
def produto_interno(vet1, vet2):

    p = sum((x * y for x, y in zip(vet1, vet2)))

    return p

def calcular_projecao(v,u):

    coef = produto_interno(v,u)/produto_interno(u,u)

    proj = [coef * u_i for u_i in u]

    return proj

def ortogonalizacao(base):

    base_ortogonal = []

    base_ortogonal.append(base[0])

    dim = len(base)

    for i in range(1,dim):

        w = base[i].copy()

        for j in range(i):

            proj = calcular_projecao(base[i], base_ortogonal[j])

            w =[w_i - proj_i for w_i,proj_i in zip(w, proj)]

        base_ortogonal.append(w)

    return base_ortogonal


base_r3 = [

    [1,0,0], 
    [1,1,0], 
    [1,1,1],

]

base_ortogonal = ortogonalizacao(base_r3)

for vetor in base_ortogonal:

    vetor = list(map(float, vetor))

    print(vetor)





```
