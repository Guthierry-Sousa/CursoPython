#### Tratamento de Erros e Exceções

```python
# Exemplo de Tratamento de erro/exceção

divisao = lambda x,y: x/y

while True:

    try:

        num1 = float(input("Informe um número(1): "))
        num2 = float(input("Informe um número(2): "))

        resultado = divisao(num1,num2)

    except ZeroDivisionError:
        print("Erro: Não é possível divisão for 0.")
        print("Tente novamente")

    except ValueError:
        print("Erro: Valor digitado é inválido (não é número).")
        print("Tente novamente")

    else:

        print(f"{num1}/{num2} = {resultado}")
        
        break

```
