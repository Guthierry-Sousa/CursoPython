##### 1 - Criando um Registro de Hóspedes

```python
quantidade_hospedes = int(input("Informe a quantidade de hóspedes (1,2,3 ou 4): "))

dados_hospedes = []

for i in range(quantidade_hospedes):

    dados = []

    nome = input("Informe seu nome: ")
    dados.append(nome)

    cpf = input("Informe seu cpf: ")
    dados.append(cpf)

    dados_hospedes.append(dados)

print("\nHóspedes cadastrados com sucesso!\n")

print("---------- Lista de Hóspedes ----------")
for hospede in dados_hospedes:
    print(hospede)
```

##### 2 - Análise de Vendas

```python
meta = 1000

vendas = []

# Preenchimento da lista de vendas
for i in range(3):
    print(f"\nFuncionário {i+1}\n")

    nome = input("Informe o nome: ")

    qtd_vendas = int(input("Informe a quantidade de vendas: "))

    dados_func = [nome, qtd_vendas]

    print(dados_func)

    vendas.append(dados_func)

print("\n----------------- Funcionários e Vendas -------------------")
print(vendas)
```

```python
# Análise da lista de vendas
for venda in vendas:

    if(venda[1] >= meta):

        print(f"\nFuncionário(a) {venda[0]} bateu a meta com {venda[1]} produtos vendidos.")
```

##### 3 - Comparação com o Ano Anterior

```python
import random

produtos = ["Notebook", "Placa De Vídeo", "Processador", "SSD", "Monitor IPS"]

vendas2024 = [random.randint(1000, 10000) for x in range(len(produtos))]
vendas2025 = [random.randint(1000, 10000) for x in range(len(produtos))]

print(f"\nVendas em 2024: {vendas2024}")
print(f"\nVendas em 2025: {vendas2025}")
```

```python
# Comparando anos de vendas

for i, produto in enumerate(produtos):

    if(vendas2025[i] > vendas2024[i]):

        procentagem_crescimento = ((vendas2025[i]/vendas2024[i]) - 1)
        
        print(f"\nProduto: {produto.capitalize()} | Crescimento: {procentagem_crescimento:.2%}")
```
