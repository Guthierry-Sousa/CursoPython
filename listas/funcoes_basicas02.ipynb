### Algumas funções básicas de listas


```python
produtos = ["Notebook", "Smartphone", "Tablet", "Fone de Ouvido", "Monitor", "Mouse Gamer", "Teclado Mecânico"]

novos_produtos = ["Cadeira Gamer", "Gabinete", "Placa de Vídeo"]
```

#### Concatenar Listas

```python
# Concatenar listas: lista1.extend(lista2) ou lista1 + lista2

produtos.extend(novos_produtos)

print(produtos)

produtos += novos_produtos

print(produtos)
```

#### Ordenação de listas (sort) - método

```python
# Ordenar listas: lista.sort

vendas = [1000, 800, 9000, 600, 5400, 3000, 2000, 100, 900, 700, 600]
vendas.sort() # Ordem crescente

print(vendas)

vendas_2 = [1000, 800, 400, 2000, 5400, 3000, 2000,]
vendas_2.sort(reverse=True) # Ordem específica 

print(vendas_2)
```

#### Ordenação de listas (sorted) - função

```python
#Ordenar Listas: lista.sorted() -> Retorna um nova lista
vendas = [1000, 800, 9000, 600, 5400, 3000, 2000, 100, 900, 700, 600]
print(f"Original: {vendas}")

vendas_ordenado = sorted(vendas)
print(f"Ordenado: {vendas_ordenado}")

# Sorted é ideal quando você não quer modificar a lista original.

nomes = ['ana', 'João', 'carla', 'Beatriz'] #lista de strings
# Ordenar ignorando maiúsculas e minúsculas
nomes_ordenado = sorted(nomes, key = str.lower)

print(f"Nomes: {nomes}")
print(f"Nomes ordenado: {nomes_ordenado}")
```

```python
# Exemplo: 

pessoas = [
    {'nome': 'João', 'idade': 30},
    {'nome': 'Maria', 'idade': 25},
    {'nome': 'Pedro', 'idade': 35}
]

#Ordenar pela idade

ordenado = sorted(pessoas, key = lambda p: p["idade"])

for p in ordenado:

    print(p)
```
