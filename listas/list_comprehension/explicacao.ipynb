```python
# List comprehension - é uma maneira de construir listas usando uma expressão concisa que pode substituir loops for e certas funções lambda. 

# [nova_expressao for item in iteralvel if condicao]

# Exemplo: Dolar para real

valores_dolares = [
    1.00, 2.50, 5.00, 10.00, 15.75, 20.00, 25.99, 30.00, 40.50,
    50.00, 75.25, 100.00, 120.40, 150.00, 175.90, 200.00,
    250.75, 300.00, 350.60, 400.00, 450.30, 500.00,
    600.00, 750.50, 800.00, 900.90, 1000.00
]

fator_conversao = 5.00

valores_reais = [d * fator_conversao for d in valores_dolares]

for d,r in zip(valores_dolares, valores_reais):

    print(f"Dolar: US${d:.2f} || Real: R${r:.2f}")
```

```python
vendas_produtos = [1500, 150, 2100, 1950]
produtos = ['vinho', 'café', 'leite', 'água']

vendas = list(zip(vendas_produtos, produtos))

#vendas_ordenada = sorted(vendas, key = lambda venda :venda[0], reverse = True)

vendas.sort(reverse = True)

vendas_produtos_ordenada = [venda[0] for venda in vendas]
produtos_ordenados = [venda[1] for venda in vendas]

print(vendas_produtos_ordenada)
print(produtos_ordenados)
```

```python
numeros = [1,2,3,4,5,6,7]

# Sem utilização de list comprehension
numeros_quadrados = []
for numero in numeros:

    numeros_quadrados.append(numero**2)

print(numeros_quadrados)

# Com utilização de list comprehension
numeros_quadrados_comp = [x**2 for x in numeros]
print(numeros_quadrados_comp)
```

```python
# Adicionando condições

import random

numeros = [random.randint(1,100) for _ in range(1000)]

numeros_pares = [x for x in numeros if x%2==0]
print(numeros_pares)

numeros_pares_maior_50 = [x for x in numeros if x%2==0 and x>50]
print(numeros_pares_maior_50)
```
