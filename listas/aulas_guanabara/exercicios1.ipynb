```python
# Faça um programa que leia 5 valores numéricos e guardeos em uma lista. No final, mostre qual foi o maior e o menor valor digitado e as suas respectivas posições na lista.

numeros = []

for i in range(5):

    numeros.append(int(input(f"Informe o {i+1} número: ")))

print("Números: ", end = '')
for n in numeros:
    print(n, end = " ")

max_value  = max(numeros) 
max_index = []
min_value = min(numeros)
min_index = []

for i,n in enumerate(numeros):

    if max_value == n:
        max_index.append(i)

    if min_value == n:
        min_index.append(i)

print(f"\nMaior: {max_value} || Indice(s): {max_index}")
print(f"Menor: {min_value} || Indice(s): {min_index}")
```

```python
# Crie um programa onde o usuário possa digitar vários valores numéricos e cadastre-os em uma lista. Caso o número já exista lá dentro, ele não sera adicionado. No final, serão exibidos todos os valores únicos digitados, em ordem crescente.

lista_numeros = []

while True:

    n = int(input("Informe um valor: "))

    if n in lista_numeros:

        print("Valor Duplicado, não será adicionado mais uma vez.")

    else:

        lista_numeros.append(n)

        print(f"{n} Adicionado.")

    op = input("Deseja continuar [s/n]? ").lower().strip()

    if op == 'n':
        break

lista_numeros.sort()
print(f"Valores digitados = {lista_numeros}")
        

    
```

```python
# Crie um programa onde o usuário possa digitar cinco valores numéricos e cadastre-os em uma lista, já na posição correta de inserção (sem usar o sort()). No final, mostre a lista ordenada na tela.

lista = []

for i in range(5):

    num = int(input("Digite um valor: "))

    if i == 0 or num > max(lista):

        lista.append(num)

        print(f"{num} foi adicionado no final da lista")

    elif num < min(lista):

        lista.insert(0,num)

        print(f"{num} foi adicionado no inicio da lista")

    else:

        for n in lista:

            if n > num:

                indice = lista.index(n)

                lista.insert(indice,num)

                print(f"{num} foi adicionado no indice {indice} da lista")

                break

print(f"\nOs valores digitados foram: {lista}\n")
```

```python
# Crie um programa que leia vários números e coloque-os em uma lista. Depois,disso, mostre:

numbers = []

while True:

    num = int(input("Digite um número: "))

    numbers.append(num)

    print(f"{num} foi adicionado na lista.")

    op = input("\nDeseja continuar [S/N]? ").strip().upper()

    if op == 'N':

        break

#a) Quantos números foram digitados;
cont_numbers = len(numbers)
print(f"\nForam digitado(s) {cont_numbers} número(s).")

#b) A lista de valores ordenada de forma decrescente;
numbers.sort(reverse = True)
print(f"\nLista em ordem decrescente: {numbers}\n")

#c) Se o valor 5 foi digitado ou não.
if 5 in numbers:
    print(f"O valor 5 foi digitado ({numbers.count(5)} vez(es)).")
else:
    print("O valor 5 não foi digitado.")
```

```python
# Crie um programa que vai ler vários números e colocar em uma lista. Depois disso, crie duas listas que vão conter apenas os valores pares e os valores ímpares digitados.

lista = []

while True:
    num = int(input("Digite um número: "))

    lista.append(num)

    print(f"{num} foi adicionado na lista.")

    op = input("\nDeseja continuar [S/N]? ").strip().upper()

    if op == 'N':

        break

lista_pares = [x for x in lista if x%2 == 0]
lista_impares = [y for y in lista if y%2 != 0]

print(f"Lista geral: {lista}")
if len(lista_pares) != 0:
    print(f"Lista pares: {lista_pares}")
else:
    print("Não foi digitado nenhum número par.")
if len(lista_impares) != 0:
    print(f"Lista ímpares: {lista_impares}")
else:
     print("Não foi digitado nenhum número ímpar.")
```

```python
# Crie um programa onde o usuário digite uma expressão qualquer que use parênteses. Seu aplicatico deverá analisar se a expressão passada está com os parênteses abertos e fechados corretamente.

expressao = input("Informe uma expressão: ").strip()

p = []

for simb in expressao:

    if simb == '(':
        p.append(simb)

    elif simb == ')':
        if len(p) > 0:
            p.pop()
            
        else:
            p.append(simb)
            break

if len(p) == 0:

    print("Expressão Válida.")

else:

    print("Expressão Inválida")
```
