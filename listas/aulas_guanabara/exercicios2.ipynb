```python
# Faça um programa que leia nome e peso de várias pessoas, guardando tudo em uma lista. No final, mostre:

pessoas = []
dados = []

while True:

    nome = input("Nome: ")
    dados.append(nome)
    
    peso = float(input("Peso(kg): "))
    dados.append(peso)

    pessoas.append(dados.copy())

    print(f"\nDados: {dados[0]} | {dados[1]}kg , CADASTRADO COM SUCESSO\n")

    dados.clear()

    op = input("Deseja continuar [s/n]? ").lower().strip()

    if op == 'n':
        break

#a) Quantas pessoas foram cadastradas;
qtd_pessoas_cadastradas = len(pessoas)
print(f"Foram cadastradas {qtd_pessoas_cadastradas} pessoas.")

#b) Uma listagem com as pessoas mais pesadas;
pesos = [p[1] for p in pessoas]
max_peso = max(pesos)
pessoas_pessadas = [p for p in pessoas if p[1] == max_peso]

print("\nPessoas mais pesadas: ")
for pessoa in pessoas_pessadas:

    print(f"Nome: {pessoa[0]} | Peso: {pessoa[1]}kg")

#c) Uma listagem com as pessoas mais leves.

min_peso = min(pesos)
pessoas_leves = [p for p in pessoas if p[1] == min_peso]

print("\nPessoas mais leves: ")
for pessoa in pessoas_leves:

    print(f"Nome: {pessoa[0]} | Peso: {pessoa[1]}kg")
```

```python
# Crie um programa onde o usuário possa digitar sete valores numéricos e cadastre-os em uma lista única que mantenha separados os valores pares e ímpares. No final, mostre os valores em ordem crescente.

numeros = [[], []]

for i in range(7):

    n = int(input(f"Informe o {i+1}o. valor: "))

    if n%2 == 0:

        numeros[0].append(n)
    
    else:

        numeros[1].append(n)


numeros[0].sort()
numeros[1].sort()

print(f"Números pares: {numeros[0]}")
print(f"Números ímpares: {numeros[1]}")
```

```python
# Crie um programa que crie uma matriz 3x3 e preencha com valores lidos pelo teclado. No final, mostre a matriz com a formatação correta.

matriz = []

for n in range(3):

    l = []

    for m in range(3):

        elemento = int(input(f"Informe o elemento ({n+1}x{m+1}): "))

        l.append(elemento)

    matriz.append(l)

print("\nMatriz: ")
for l in matriz:
    print(l)
```

```python
# Aprimore o desafio anterior, mostrando: 
#a) A soma de todos os valores pares digitados;
soma_pares = 0
for l in matriz:
    for i in l:
        if i%2 == 0:
            soma_pares += i

print(f"Soma dos elementos pares: {soma_pares}")

#b) A soma dos valores da terceira coluna;
soma_terceira_coluna = 0
for l in matriz:
    soma_terceira_coluna += l[2]

print(f"Soma dos elementos da terceira coluna: {soma_terceira_coluna}")

#c) O maior valor da segunda linha.
max_valor_segunda_linha = max(matriz[1])

print(f"O maior valor da segunda linha é: {max_valor_segunda_linha}")
```

```python
# Faça um programa que ajude um jogador da MEGA SENA a criar palpites. O programa vai perguntar quantos jogos serão gerados e vai sortear 6 números entre 1 e 60 para cada jogo, cadastrando tudo em uma lista composta.
import random
import time

qtd_jogos = int(input("Informe a quantidade de jogos: "))

lista_jogos = []

for i in range(qtd_jogos):

    numeros_sorteados = random.sample(range(1,61), 6)

    numeros_sorteados.sort()

    lista_jogos.append(numeros_sorteados[:])

    numeros_sorteados.clear()

print(f"Sorteando {qtd_jogos} jogos ...")
for i,jogo in enumerate(lista_jogos):

    print(f"Jogo {i+1}: {jogo}")

    time.sleep(1)

print("BOA SORTE!")
    
```

```python
# Crie um programa que leia nome e duas notas de vários alunos e guarde tudo em uma lista composta. No final, mostre um boletim contendo a média de cada um e permita que o usuário possa mostrar as notas de cada aluno individualmente.

dados_alunos = []

while True:
    aluno = []
    notas = []

    nome = input("Nome do aluno: ").strip().upper()

    nota1 = float(input("Informe a nota 1: "))
    nota2 = float(input("Informe a nota 2: "))

    notas.append(nota1)
    notas.append(nota2)

    media = sum(notas) / len(notas)
    notas.append(media)

    aluno.append(nome)
    aluno.append(notas)

    dados_alunos.append(aluno)

    op = input("Deseja continuar [s/n]? ").lower().strip()
    if op == 'n':
        break

print("\nBOLETIM DOS ALUNOS")
print("-" * 30)

for i, aluno in enumerate(dados_alunos):
    print(f"{i} - Nome: {aluno[0]} | Média: {aluno[1][2]:.2f}")

while True:
    n = int(input("\nInforme o número do aluno (999 para parar): "))

    if n == 999:
        break

    if n < 0 or n >= len(dados_alunos):
        print("Aluno inválido!")
        continue

    print(f"Nome: {dados_alunos[n][0]} | Notas: {dados_alunos[n][1][0:2]} | Média: {dados_alunos[n][1][2]:.2f}")

print("\nPROGRAMA FINALIZADO!\n")




```
