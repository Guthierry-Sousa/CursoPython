### Algumas Funções Básicas de Lista

```python
produtos = ["celular", "tv", "monitor", "processador", "placa de vídeo", "cadeira gamer", "placa-mãe"]
```

```python
# Tamanho da Lista: Quantos itens a lista possui

qtd_produtos = len(produtos)

print(f"Temos {qtd_produtos} produtos.")
```

```python
# Maior(max) e Menor(min) valor

vendas = [1000, 2000, 1500, 17500, 5000, 700, 450]

menos_vendido = min(vendas)

mais_vendido = max(vendas)

print(f"O Produto mais vendido teve {mais_vendido} unidades vendidas.")

print(f"O Produto menos vendido teve {menos_vendido} unidades vendidas.")
```

```python
# Exemplo: 

# Lista de produtos de tecnologia
produtos = ["Notebook", "Smartphone", "Tablet", "Fone de Ouvido", "Monitor", "Mouse Gamer", "Teclado Mecânico"]

# Lista de quantidades vendidas (baseada no índice dos produtos)
quantidades_vendas = [120, 300, 150, 500, 200, 350, 180]

menos_vendido = min(quantidades_vendas)

mais_vendido = max(quantidades_vendas)

indice_min = quantidades_vendas.index(menos_vendido)

indice_max = quantidades_vendas.index(mais_vendido)

# Produto mais vendido

print(f"O produto mais vendido foi {produtos[indice_max]} com {mais_vendido} unidades vendidas. ")

# Produto menos vendido

print(f"O produto menos vendido foi {produtos[indice_min]} com {menos_vendido} unidades vendidas. ")
```
