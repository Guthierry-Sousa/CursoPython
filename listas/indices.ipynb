### Listas em Python: índices

```python
# Como descobrir o índice de um item em uma lista?

# - i = lista.index('item')

produtos = ["celular", "tv", "monitor", "processador", "placa de vídeo"]

vendas = [1000, 2000, 1500, 17500, 5000]

i = produtos.index("processador")

print(i) # 0

qtd_vendas = vendas[i]

print(f"Quantidade de vendas do produto '{produtos[i]}' = {qtd_vendas} unidades")
```

```python
# Exercício: Consulta de Estoque
produtos = ["celular", "tv", "monitor", "processador", "placa de vídeo"]

estoque = [1000, 2000, 1500, 17500, 5000]

nome_produto = input("Informe o nome do produto: ").lower()

if nome_produto in produtos:

    i = produtos.index(nome_produto)

    qtd_estoque = estoque[i]

    print(f"'{nome_produto}' possui {qtd_estoque} unidades.")

else:

    print(f"Não possuimos o produto '{nome_produto}'.")
```
