### Exercicio 01:

```python
meses = [
    "Janeiro", "Fevereiro", "Março", "Abril", "Maio", "Junho",
    "Julho", "Agosto", "Setembro", "Outubro", "Novembro", "Dezembro"
]

vendas_primeiro_semestre = [120, 95, 110, 130, 150, 140]

vendas_segundo_semestre = [160, 170, 180, 200, 100, 400]
```

```python
vendas_meses = vendas_primeiro_semestre + vendas_segundo_semestre

print(vendas_meses)

maior_valor = max(vendas_meses)

menor_valor = min(vendas_meses)

index_max = vendas_meses.index(maior_valor)

index_min = vendas_meses.index(menor_valor)

best_mes = meses[index_max]

pior_mes = meses[index_min]

print(f"O melhor mês do ano foi {best_mes} com {maior_valor} vendas")
print(f"O pior mês do ano foi {pior_mes} com {menor_valor} vendas")

```

```python
# Calculando Faturamento total:

faturamento_total = sum(vendas_meses)

fat_melhor_mes = (maior_valor/faturamento_total) * 100

print(f"Faturamento Total = {faturamento_total}")

print(f"Faturamento do melhor mes em relação ao total foi de {fat_melhor_mes:.2f}%")
```

```python
# Top 3 valores de vendas

vendas_ordenado = sorted(vendas_meses)

top3 = []

for i in range(3):

    item = vendas_ordenado.pop()

    top3.append(item)

print(top3)
```
