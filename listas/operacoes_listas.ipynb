### Adicionar e Remover itens em uma lista

```python
# Adicionar item: lista.append(item) - Adiciona no final da lista

# Remover item: pop (retorna o item removido) e remove (apenas remove o item)

nomes = ['maria', 'elis', 'elen', 'guthierry', 'ramon']
```

```python
item_removido = nomes.pop() # Por padrão remove o ultimo item

nomes.remove('elen') # Remove o item especificado (pode causar erros caso o item não exista)

print(item_removido)

print(nomes)
```

```python
nomes.append("messi") # Por padrão adiciona no final da lista

nomes.insert(0, "cristiano ronaldo") # Adiciona o item no indice especificado 

print(nomes)
```

#### Tratando erros possíveis

```python
# Utilizando if

produtos = ["celular", "tv", "monitor", "processador", "placa de vídeo"]

item = input("Informe o item que você deseja remover: ").strip().lower()

if item in produtos:

    produtos.remove(item)

    print(f"'{item}' removido do estoque!")

else:

    print(f"'{item}' ja não existe no estoque!")

print(f"Produtos: {produtos}")
```

```python
# Utilizando try...except

try:
    item = input("Informe o item que você deseja remover: ").strip().lower()

    produtos.remove(item)

    print(f"'{item}' removido do estoque!")

except:
    
    print(f"'{item}' ja não existe no estoque!")

print(f"Produtos: {produtos}")
```
