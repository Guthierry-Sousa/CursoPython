#### Iterable
Qualquer objeto que expõe uma interface de iteração, permitindo que seus elementos sejam acessados sequencialmente.

```python
# Listas

numbers = [1,2,3,4,5]

for num in numbers:
    print(num, end = " ")
```

```python
# Tuplas

coords = (10,20,30)

for c in coords:
    print(c)
```

```python
# Strings

texto = "guthierry"

for ch in texto:
    print(ch, end = "|")

# Cada elemento é uma string de comprimento 1
# Não são caracteres no sentido de C, mas objetos str
```

```python
# Conjuntos - não pode ter valores duplicados

s = {1,3,5,7,9}

for x in s:
    print(x)

# Iterável sem ordem definida
s_lista = list(s) #Ordem não garantida
print(s_lista)
```

```python
# Dicionários

d = {'a':1, 'b':2, 'c':3}

for k in d:
    print(k)

print("------------")

for v in d.values():
    print(v)

print("------------")

for k,v in d.items():
    print(k,v)

# keys(), values() e items() retornam estruturas iteráveis dinâmicas
```

```python
# Range

r = range(1,100)

for x in r:
    print(x)

# Não armazena todos os valores na memória
# Calcula cada valor sob demanda
```

```python
# Resultado de funções built-in que retornam iterables

#Enumerate
nomes = ['Elis', 'Guthierry']

for i, name in enumerate(nomes):
    print(f"Index: {i} | Name: {name}")

print("------------------")

#Zip
a = [1,2]
b = [3,4]

for x,y in zip(a,b):
    print(x,y)
```
