```python
niveis_criticos_co2 = {}
for estado, niveis in niveis_co2.items():

    media_niveis_co2 = sum(niveis)/len(niveis)

    if media_niveis_co2 > 450:

        print(f"Estado: {estado} | Nível médio de CO2: {media_niveis_co2} (CRÍTICO).")

        niveis_criticos_co2[estado] = media_niveis_co2

    else: 

        print(f"Estado: {estado} | Nível médio de CO2: {media_niveis_co2} (NORMAL).")

```
