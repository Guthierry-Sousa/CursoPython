```python
# Exercício 01: Faça um programa que leia nome e média de uma aluno, guardando em um dicionário. No final, mostre o conteúdo da estrutura na tela.

dados_alunos = []
aluno = {}

while True:

    aluno['Nome'] = input("Aluno (nome): ")
    aluno['Média'] = float(input("Média: "))

    if aluno['Média'] >= 7.0:

        aluno['Situação'] = "Aprovado"

    else:

        aluno['Situação'] = "Reprovado"

    dados_alunos.append(aluno.copy())

    op = input("Deseja continuar[s/n]? ").strip().upper()

    if op == 'N':

        break

for i,aluno in enumerate(dados_alunos):

    print(f"\nALUNO: {i+1}")

    for k,v in aluno.items():

        print(f"{k}:{v}")
```

```python
# Exercício 02: Crie um programa onde 4 jogadores joguem um dado e tenham resultados aleatórios. Guarde esses resultados em um dicionário. No final coloque esse dicionário em ordem, sabendo que o vencedor tirou o maior número no dado.

import random
import time

jogadores = {}

for i in range(4):
    jogadores[f'Jogador {i+1}'] = random.randint(1,6)

print("Valores sorteados: ")
for k,v in jogadores.items():

    print(f"O {k} tirou {v}")
    time.sleep(1)

# Ordenação de dicionários
#jogadores = list(jogadores.items())
#jogadores.sort(key = lambda jogador: jogador[1], reverse = True)
jogadores = sorted(jogadores.items(), key = lambda item: item[1], reverse = True)
jogadores = dict(jogadores)

print("Ranking dos jogadores: ")
cont = 0
for k,v in jogadores.items():

    print(f"{cont+1}o. Lugar: {k} | Valor sorteado: {v}")
    time.sleep(1)

    cont += 1


```

```python
# Exercício 03: Crie um programa que leia nome, ano de nascimento e carteira de trabalho e cadastre-os (com idade) em um dicionário, se por o CTPS for diferente de 0, o dicionário receberá tambem o ano de contratação e o salário.

import time

dados = {}

dados['Nome'] = input("Informe seu nome: ").strip().capitalize()

data_nascimento = int(input("Informe sua data de nascimento: "))
ano_atual = time.gmtime().tm_year #datetime.now().year
dados['Idade'] = ano_atual - data_nascimento

dados['CTPS'] = int(input("Informe o número da carteira de trabalho (0 se não tiver): "))

if dados['CTPS'] != 0:

    dados['Contratação'] = int(input("Informe seu ano de contratação: "))
    dados['Salário'] = float(input("Informe seu salário: "))

    ano_aposentadoria = ano_atual - dados['Contratação']

    dados['Aposentadoria'] = ano_aposentadoria

for k, v in dados.items():

    print(f"{k}: {v}")


```

```python
# Exercício 04: Crie um programa que gerencie o aproveitamento de um jogador de futebol. O programa vai ler o nome do jogador e quantas partidas ele jogou. Depois vai ler a quantidade de gols feito em cada partida.

dados_jogador = {}

gols = []

dados_jogador['Nome'] = input("Informe o nome do jogador: ")

qtd_jogos = int(input("Informe a quantidade de jogos: "))
for i in range(1,qtd_jogos+1):

    gols.append(int(input(f"Quantos gols na partida {i}: ")))

dados_jogador['Gols'] = gols
dados_jogador['Total'] = sum(gols)

print("-"*30)
print(dados_jogador)
print("-"*30)

print(f"\nNome do jogador: {dados_jogador['Nome']}")
print(f"Gols por cada partida jogada: {dados_jogador['Gols']}")
print(f"Total de gols: {dados_jogador['Total']}\n")

print("-"*30)

print(f"O jogador {dados_jogador['Nome']} jogou {qtd_jogos} partidas.")

for i, qtd_gols in enumerate(dados_jogador['Gols']):

    print(f"Partida {i+1} : {qtd_gols} gols.")



```

```python
# Crie um programa que leia nome, sexo e idade de várias pessoas. No final, mostre:

dados_pessoas = []

while True:

    pessoa = {}

    pessoa['Nome'] = input("Nome: ")
    pessoa['Sexo'] = input("Sexo (M/F): ").strip().upper()
    pessoa['Idade'] = int(input("Idade: "))

    dados_pessoas.append(pessoa)

    op = input("Deseja continuar[s/n]? ").strip().upper()

    if op == 'N':

        break

#a) Quantas pessoas foram cadastradas;
qtd_pessoas_cadastradas = len(dados_pessoas)
print(f"\nQuantidade de pessoas cadastradas: {qtd_pessoas_cadastradas}")

#b) A média de idade do grupo;
soma_idades = sum(pessoa['Idade'] for pessoa in dados_pessoas)
media_idade = soma_idades/qtd_pessoas_cadastradas
print(f"\nA média de idade do grupo é {round(media_idade)} anos.")

#c) Uma lista com todas as mulheres;
mulheres = [pessoa['Nome'] for pessoa in dados_pessoas if pessoa['Sexo'] == 'F']
print(f"\nMulheres cadastradas = {','.join(mulheres)}")

#d) Uma lista com todas as pessoas com idade acima da média.
pessoas_acima_media = [pessoa['Nome'] for pessoa in dados_pessoas if pessoa['Idade'] > media_idade]
print(f"\nPessoas com idade acima da média: {' , '.join(pessoas_acima_media)}")
```
