#### Métodos Úteis em dicionários

```python
# Items()

vendas_tecnologia = {
    'Notebook': 120,
    'Smartphone': 350,
    'Tablet': 95,
    'Monitor': 180,
    'Teclado': 210,
    'Mouse': 260,
    'Impressora': 75,
    'Scanner': 40,
    'HD Externo': 130,
    'SSD': 220,
    'Pendrive': 300,
    'Roteador': 160,
    'Switch de Rede': 55,
    'Placa de Video': 90,
    'Placa Mae': 110,
    'Processador': 105,
    'Memoria RAM': 240,
    'Fonte de Alimentacao': 85,
    'Webcam': 140,
    'Headset': 190,
    'Smartwatch': 125,
    'Caixa de Som': 170
}

# Exibir itens

for itens in vendas_tecnologia.items():

    print(itens)

# Produto mais vendido

#mais_vendido = max(vendas_tecnologia, key = vendas_tecnologia.get)
mais_vendido = max(vendas_tecnologia, key = lambda chave: vendas_tecnologia.get(chave))

print(f"\nProduto mais vendido: {mais_vendido}\n")

# Produtos que venderam mais de 200 unidades

for produto, unidades_vendas in vendas_tecnologia.items():

    if unidades_vendas > 200:

        print(f"Produto: {produto.upper()} | Vendas: {unidades_vendas}")
```

#### Listas importantes a partir do Dicionário

```python
# Método .keys() - 'Lista' com as chaves do dicionário

print(list(vendas_tecnologia.keys()))

# Método .values() - 'Lista' com os valores do dicionário
print(list(vendas_tecnologia.values()))

# Organizando o dicionário em ordem alfabética
vendas_tecnologia = dict(sorted(vendas_tecnologia.items()))

print(vendas_tecnologia)
```

#### Transformando Listas em Dicionários e Function zip

```python
produtos = [
    'Notebook',
    'Smartphone',
    'Tablet',
    'Monitor',
    'Teclado',
    'Mouse',
    'Impressora',
    'SSD',
    'HD Externo',
    'Roteador'
]

quantidade_vendas = [
    120,
    350,
    95,
    180,
    210,
    260,
    75,
    220,
    130,
    160
]

# Zip cria um iterador,e iteradores só podem ser consumidos uma vez
dados_vendas = zip(produtos, quantidade_vendas) #Lista de tuplas (produto, quantidade de vendas)

# Criando um dicionário: 
dados_vendas_dicionario = dict(dados_vendas)

print(dados_vendas_dicionario)
```
