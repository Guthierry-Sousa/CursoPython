##### WebBrowser

```python
# WebBrowser - fornece uma interface de alto nível para abrir e controlar navegadores da web a partir de scripts Python

# Abrindo um site na internet

import webbrowser as wb

def abrir_site(url):

    try:
        wb.open(url)
    except:
        print("Erro: Não foi possível abrir o site informado")

abrir_site('https://www5.usp.br')
abrir_site('https://unicamp.br')
```

##### Time

```python
# Time - serve para manipular e obter informações sobre tempo.

import time

# Marco Zero (EPOCH) = 1 de janeiro de 1970 as 00:00:00

# time() retorna quantos segundos se passaram desde a EPOCH
segundos_hoje = time.time()
print(segundos_hoje)
```

```python
# ctime() retorna a data em string no formato UTC
data_hoje = time.ctime().split()

print(data_hoje)

dia_semana = data_hoje[0]
mes = data_hoje[1]
dia = data_hoje[2]
horario = data_hoje[3]
ano = data_hoje[4]

print(f"Dia da Semana : {dia_semana} | Mês: {mes} | Dia (data): {dia} | Horário: {horario} | Ano: {ano}")
```

```python
# Exemplo: Temporizar um programa
import random as rd

tempo_inicial = time.time()

numeros = [rd.randint(1,100) for _ in range(100000000)]
valores_dobrados = [x*2 for x in numeros]

tempo_final = time.time()

tempo_total = tempo_final - tempo_inicial

print(f"O programa demorou {tempo_total:.2f}s para ser executado")
```

```python
# sleep() faz o código esperar alguns segundos

def contador(n):

    for i in range(n+1):

        print(i)

        time.sleep(1)


contador(10)
```

```python
# gmtime() Pega informações de dias, horas, segundos, minutos,

data_atual = time.gmtime()

print(data_atual)

mes_atual = time.gmtime().tm_mon #data_atual.tm_mon
print(mes_atual)

ano_atual = data_atual.tm_year
print(ano_atual)

dia_atual = data_atual.tm_mday
print(dia_atual)

print(f"Data: {dia_atual}/{mes_atual}/{ano_atual}")
```

#### Matplotlib

```python
import matplotlib.pyplot as plt

vendas = [1500, 100, 1000, 600, 2000, 900]
meses = ['jan', 'fev', 'mar', 'abril', 'maio', 'julho']

plt.title("Vendas X Meses")
plt.plot(meses, vendas, '--r')
plt.xlabel('Meses (1 - Semestre)')
plt.ylabel('Vendas')
plt.grid()
plt.show()
```

##### NumPy + Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np

def s(x):

    return np.sin(x)

def c(x):

    return np.cos(x)

x = np.arange(0, np.pi*5, 0.1)
y_s = s(x)
y_c = c(x)

plt.plot(x, y_s, 'b-', label = "f(x) = Sin(x)")
plt.plot(x, y_c, 'r-',label = "f(x) = Cos(x)")
plt.legend()
plt.show()
```

```python
# Grafico de barras

import matplotlib.pyplot as plt

vendas = [1500, 100, 1000, 600, 2000, 900]
meses = ['jan', 'fev', 'mar', 'abril', 'maio', 'julho']

plt.title("Vendas X Meses")
plt.bar(meses, vendas)
plt.xlabel('Meses (1 - Semestre)')
plt.ylabel('Vendas')
plt.grid()
plt.show()
```
