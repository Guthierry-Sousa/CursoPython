```python
import moedas
# Crie um módulo que tenha as funções incorporadas aumentar(), diminuir(), dobrar() e metade(). Faça também um programa que importe esse módulo e use algumas dessas funções.

p = float(input("Informe o preço: "))

metade = moedas.metade(p)
dobro = moedas.dobrar(p)
preco_aumentado = moedas.aumentar(p, 10)
preco_diminuido = moedas.diminuir(p, 13)


print(f"\nA metade de {p} é {metade}\n")
print(f"\nO dobro de {p} é {dobro}\n")
print(f"\nAumentando {p} em 10% = {preco_aumentado}\n")
print(f"\nDiminuindo {p} em 13% = {preco_diminuido}\n")
```

```python
# Adapte o código anterior, criando uma função adicional chamada formatar() que consiga mostrar os valores como um valor monetário formatado.

print(f"\nA metade de {moedas.formatar(p)} é {metade}\n")
print(f"\nO dobro de {moedas.formatar(p)} é {dobro}\n")
print(f"\nAumentando {moedas.formatar(p)} em 10% = {preco_aumentado}\n")
print(f"\nDiminuindo {moedas.formatar(p)} em 13% = {preco_diminuido}\n")
```

```python
# Modifique as funções que foram criadas anteriormente para que elas aceitem um parâmetro a mais, informando se o valor retornado por elas vai ser ou não formatado pela função formatar().

p = float(input("Informe o preço: "))

metade = moedas.metade(p, True)
dobro = moedas.dobrar(p, False)
preco_aumentado = moedas.aumentar(p, 10, False)
preco_diminuido = moedas.diminuir(p, 13, True)

print(f"\nA metade de {moedas.formatar(p)} é {metade}\n")
print(f"\nO dobro de {moedas.formatar(p)} é {dobro}\n")
print(f"\nAumentando {moedas.formatar(p)} em 10% = {preco_aumentado}\n")
print(f"\nDiminuindo {moedas.formatar(p)} em 13% = {preco_diminuido}\n")
```
