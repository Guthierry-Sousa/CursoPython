```python
# Exercício 01:  Crie um programa que tenha uma tupla totalmente preenchida com uma contagem por extenso, de zero até 20. Seu programa deverá ler um número pelo teclado (entre 0 a 20) e mostrá-lo por extenso.

contagem = ('zero', 'um', 'dois', 'três', 'quatro', 'cinco', 'seis', 'sete', 'oito', 'nove', 'dez', 'onze', 'doze', 'treze', 'quatorze', 'quinze', 'dezesseis', 'dezessete', 'dezoito', 'dezenove', 'vinte')

while True:
    try:
        num = int(input("Informe um número de 0 até 20: "))
        
    except: 
        print("Digite apenas números!")
    else:

        if 0<=num<=20:

            print(f"Número: {contagem[num]}({num})")
            break

        else:
    
            print("Você digitou um número fora do intervalo de 0 até 20.")

```

```python
# Exercício 02: Crie uma tupla preenchida com os 20 primeiros colocados do Campeonato Brasileiro de Futebol, na ordem de colocação. Depois mostre:

times_brasileirao_2025 = (
    "Flamengo",
    "Sociedade Esportiva Palmeiras",
    "Cruzeiro Esporte Clube",
    "Mirassol Futebol Clube",
    "Fluminense Football Club",
    "Botafogo de Futebol e Regatas",
    "Esporte Clube Bahia",
    "São Paulo Futebol Clube",
    "Grêmio Foot-Ball Porto Alegrense",
    "Red Bull Bragantino",
    "Clube Atlético Mineiro",
    "Santos Futebol Clube",
    "Sport Club Corinthians Paulista",
    "Club de Regatas Vasco da Gama",
    "Esporte Clube Vitória",
    "Sport Club Internacional",
    "Ceará Sporting Club",
    "Fortaleza Esporte Clube",
    "Esporte Clube Juventude",
    "Sport Club do Recife"
)

# a) Apenas os 5 primeiros colocados;
top_5 = times_brasileirao_2025[:5]
print(f"Top 5: {top_5}\n")

# b) Os últimos 4 colocados;
ultimos_4 = times_brasileirao_2025[-4:]
print(f"Z4: {ultimos_4}\n")

# c) Uma lista com os times em ordem alfabética;
times_ordem_alfabetica = sorted(times_brasileirao_2025)
print(f"Times em ordem alfabética: {times_ordem_alfabetica}\n")

# d) Em que posição na tabela está o Vasco.
for time in times_brasileirao_2025:

    if "VASCO" in time.upper():

        pos_vasco = times_brasileirao_2025.index(time) + 1

        print(f"A posição do {time} foi: {pos_vasco} colocado.")

        break


```

```python
# Exercício 03: Crie um programa que vai gerar cinco números aleatórios e coloque eles em uma tupla. Depois disso, mostre a listagem de números gerados e tambem indique o menor e o maior valor que estão na tupla.

import random

numeros_aleatorios = tuple(random.randint(1,10) for _ in range(5))

print(f"Números gerados: {numeros_aleatorios}")

max_value = max(numeros_aleatorios)
min_value = min(numeros_aleatorios)

print(f"Maior: {max_value} | Menor: {min_value}")
```

```python
# Exercícios 04: Desenvolva um programa que leia quatro valores pelo teclado e guarde-os em uma tupla. No final, mostre:

numeros_digitados = []

for i in range(5):

    num = int(input(f"Informe o {i+1} número: "))

    numeros_digitados.append(num)

numeros_digitados = tuple(numeros_digitados)
print(f"Números Digitados: {numeros_digitados}")

#a) Quantas vezes o valor 9 apareceu;
cont_9 = numeros_digitados.count(9)
print(f"O número 9 apareceu {cont_9} vezes.\n")

#b) Em que posição foi digitado o primeiro valor 3;
try:
    posicao_3 = numeros_digitados.index(3) + 1
except:
    print("O número 3 não foi digitado.")
else:
    print(f"A posição do primeiro número 3 é: {posicao_3}")

#c) Quais foram os números pares.
pares = tuple(n for n in numeros_digitados if n % 2 == 0)
if len(pares) == 0:

    print("Nenhum número par foi digitado.")

else:

    print(f"Os números pares foram: {pares}")
```

```python
# Exercício 06: Crie um programa qu tenha uma tupla com várias palavras. Depois disso, você deve mostrar, para cada palavra, quais são as suas vogais.

palavras = (
    "computacao",
    "algoritmo",
    "programacao",
    "estrutura",
    "dados",
    "inteligencia",
    "artificial",
    "seguranca",
    "informacao",
    "criptografia",
    "rede",
    "sistema",
    "hardware",
    "software",
    "desenvolvimento",
    "analise",
    "codigo",
    "python",
    "linux",
    "ciberseguranca"
)

vogais = 'aeiou'

for palavra in palavras:

    print(f"\nA palavra {palavra.upper()} possui as vogais: ", end = " ")

    for letra in palavra:

        if letra in vogais:

            print(letra, end = " ")

    
```
